# Create synthetic Forward Configs for `Forward Simulation.py`

### Simple Movement
- Uniform Angles
- Constant Position
- Constant Rotation Axis

```python
import json
import numpy as np
import os

output_dir = "../../../Configs/Experiment Configs/Simple Propergation 23_07_2025"
file_name = "simple_movement.json"

output_path = os.path.join(output_dir,file_name) 
os.makedirs(output_dir, exist_ok=True)

timesteps = 485

# Values independent of best pose
augmentation_dict={
    "field": {
          "crop": {"x_min":125, "y_min":125, "x_max":-125, "y_max":-125},
    },
	"amp": {},
    "phase": {}		
}

movement_dict={
      "time_steps": timesteps,
      "unit": "um",
      "Positions":[],
      "Offset": [
           { "offset": [0, 0, 0], "time": 0 } 
      ],
      "Rotation":[]
}

# Rotation - Manual to circumvent Slerp interpolation
for t in range(timesteps):
    angle = t/timesteps * 360
    pos_dict = {"pos": [25.0,25.0,25.0], "time": t}
    rot_dict = {"axis": [1.0, 0.0, 0.0], "theta": angle, "time": t}

    movement_dict["Positions"].append(pos_dict)
    movement_dict["Rotation"].append(rot_dict)


# Forward Config
data = {
      "Augmentations": augmentation_dict,
      "Movement": movement_dict
}

# Save Forward Config
with open(output_path, 'w') as f:
    json.dump(data, f, indent=2)
```

### Non Uniform Angles
Fixed Axis<br>
Fixed Position

```python
import numpy as np
import matplotlib.pyplot as plt

# in rads

x = np.linspace(0,2*np.pi, 485)

# uniform
uniform_angles = x

# non uniform
amp=0.25
k=3
non_uniform_angles = x + amp * np.sin(k*x)


# rad --> deg
x = np.rad2deg(x)
uniform_angles = np.rad2deg(uniform_angles)
non_uniform_angles = np.rad2deg(non_uniform_angles)

plt.plot(x,uniform_angles, ls="--", color="r", label="Uniform Angles")
plt.plot(x,non_uniform_angles, ls="-", color="b", label="Non-Uniform Angles")

plt.legend()
plt.grid()
plt.show()
```

### Move Object
Uniform Angle <br>
Fixed Axis

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0,2*np.pi, 485, endpoint=False)

r = 5
positions = np.array([25,25,25]) + r * np.array([np.cos(x),np.sin(x),np.zeros_like(x)]).T


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

center = np.array([25,25,25])


ax.scatter(center[0],center[1],center[2], color="r", label="Center")
ax.scatter(positions[:, 0], positions[:, 1], positions[:, 2], color="b", marker=".", label=f"Object Trajectory \n r={r}")

ax.set_xlim(0, 50)
ax.set_ylim(0, 50)
ax.set_zlim(0, 50)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.legend()
plt.tight_layout()
plt.show()
```
