### Creates a new .mat file from the original Coded WFS frames. 
Add temporal blur to the OPD

```python
original_data_file = "../../Data/Raw Data/HEK Cells/Recorded Frames/raw_coded_wfs.mat"

temp_blur_data_file = "../../Data/Raw Data/HEK Cells/Recorded Frames/raw_coded_wfs_temp_blur.mat"
```

```python
import scipy

# load Data
data = scipy.io.loadmat(original_data_file)
print(data.keys())
```

```python
# Temporal blur
sigma = 5
opd = data['opd']
amp = data['amp']

opd = scipy.ndimage.gaussian_filter(opd, sigma, axes=0)
amp = scipy.ndimage.gaussian_filter(amp, sigma, axes=0)
```

```python
### Create new .mat file ###
new_data = {
    "amp": amp,
    "opd": opd,
    "opd_units": data['opd_units'],
    "px_size": data['px_size'],
    "temp_blur_sigma": sigma
}

scipy.io.savemat(temp_blur_data_file, new_data)
```
