# Convert Pose Opt output file to input file for Recon Opt
Take best settings from pose opt, convert into pose file for recon opt.
<br>
- Input: PoseOpt/run/best_settings.json <br>
- Output: Configs/Recon/Poses/poses.json

### DHM Poses
Best found poses for DHM recorded frames

```python
pose_opt_file = "../../Outputs/PoseOpt/HEK Cells/DHM/DHM PoseOpt (2)/Best Settings/best_settings.json"
recon_opt_file ="../../Configs/Sample Configs/ReconOpt/Pose Files/dhm_poses.json"
```

```python
import json

with open(pose_opt_file, 'r') as f:
    pose_opt = json.load(f)

recon_opt = {}

for frame, info in pose_opt.items():


    idx = frame.split(' ')[-1]
    idx = int(idx)

    unit = info["unit"]
    position = info["Position"]
    axis = info["Axis"]
    angle = info["Theta"]

    recon_opt[frame] = {
        "idx": idx,
        "unit": unit,
        "Position": position,
        "Axis": axis,
        "Angle": angle
    } 



# Save Forward Config
with open(recon_opt_file, 'w') as f:
    json.dump(recon_opt, f, indent=2)
```
